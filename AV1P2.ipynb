**1. INTRODUÇÃO**

Nesta implementação o foco era desenvolver um sistema de classificação de imagens de pragas de juta, contendo um total de 17 features e utilizando redes neurais convolucionais, conhecidas também como CNN, e algoritmos de classificação. A metodologia inclui a extração de características com o modelo VGG16, a redução de dimensionalidade com PCA (Análise de Componentes Principais) e a aplicação de classificadores como `NearestCentroid` e `KNeighborsClassifier` (k-NN) usando diferentes métricas de distância (Euclidiana, Manhattan e Mahalanobis).

O código começa importando bibliotecas essenciais como `tensorflow` para manipulação de redes neurais, `sklearn` para algoritmos de aprendizado de máquina e `mlxtend` para visualização de regiões de decisão. Em seguida, define os diretórios para os conjuntos de dados de treinamento, validação e teste.

Utiliza-se `ImageDataGenerator` para pré-processamento e geração de lotes de imagens. O modelo VGG16 pré-treinado é carregado, sem a camada totalmente conectada final, e a saída da rede é achatada para ser usada como vetor de características. Uma função `extract_features` é definida para extrair características das imagens, aplicando padding para garantir que todas tenham o mesmo tamanho.

As características são extraídas dos conjuntos de dados de treinamento, validação e teste usando o modelo VGG16. Em seguida, os dados são padronizados com `StandardScaler` para ter média zero e desvio padrão unitário. Aplica-se PCA para reduzir a dimensionalidade dos dados para 50 componentes principais para treinamento e para 2 componentes principais para visualização.

Calcula-se a matriz de covariância e sua inversa para a métrica de Mahalanobis. Diferentes métricas e valores de `k` são usados para treinar os classificadores `NearestCentroid` e `k-NN`. Para cada combinação, são geradas predições sobre o conjunto de teste, calculadas as matrizes de confusão e as acurácias.

Os resultados são visualizados através de subplots, que incluem regiões de decisão dos classificadores `NearestCentroid` e `k-NN`, e matrizes de confusão para ambos, com as acurácias nos títulos. As visualizações são organizadas em um layout de 3x2 para facilitar a comparação.

Finalmente, o código calcula e imprime estatísticas descritivas (mínimo, máximo, média e desvio padrão) das acurácias dos classificadores `NearestCentroid` e `k-NN`.

Este projeto oferece uma abordagem abrangente para a classificação de imagens utilizando técnicas de aprendizado profundo e aprendizado de máquina. A extração de características com um modelo pré-treinado, seguida da redução de dimensionalidade e classificação com diferentes métricas, permite uma análise detalhada e visualização dos resultados, proporcionando insights valiosos sobre a eficácia dos classificadores em diferentes contextos.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestCentroid, KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from scipy.spatial.distance import mahalanobis
from mlxtend.plotting import plot_decision_regions
from sklearn.decomposition import PCA
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Flatten
import tensorflow as tf
import cv2

# Classe customizada para a métrica Mahalanobis no classificador Nearest Centroid
class NearestCentroidMahalanobis(NearestCentroid):
    def __init__(self):
        super().__init__()

    def fit(self, X, y):
        super().fit(X, y)
        self.cov_ = np.cov(X, rowvar=False)
        self.inv_covmat_ = np.linalg.inv(self.cov_)

    def predict(self, X):
        return np.array([np.argmin([mahalanobis(x, c, self.inv_covmat_) for c in self.centroids_]) for x in X])

# Definir caminhos
base_dir = '/content/drive/MyDrive/Colab Notebooks/Jute_Pest_Dataset'
train_dir = f'{base_dir}/train'
val_dir = f'{base_dir}/val'
test_dir = f'{base_dir}/test'

# Dimensões das imagens
img_height, img_width = 300, 300

# Geradores de dados
train_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
val_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
test_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)

train_generator = train_datagen.flow_from_directory(train_dir, target_size=(img_height, img_width), batch_size=32, class_mode='sparse')
val_generator = val_datagen.flow_from_directory(val_dir, target_size=(img_height, img_width), batch_size=32, class_mode='sparse')
test_generator = test_datagen.flow_from_directory(test_dir, target_size=(img_height, img_width), batch_size=32, class_mode='sparse')

# Carregar modelo VGG16
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(img_height, img_width, 3))
model = Model(inputs=base_model.input, outputs=Flatten()(base_model.output))

# Função para preencher imagens
def pad_image(image, target_height, target_width):
    original_height, original_width = image.shape[:2]
    delta_width = target_width - original_width
    delta_height = target_height - original_height
    top, bottom = delta_height // 2, delta_height - (delta_height // 2)
    left, right = delta_width // 2, delta_width - (delta_width // 2)
    color = [0, 0, 0]
    padded_image = cv2.copyMakeBorder(image, top, bottom, left, right, cv2.BORDER_CONSTANT, value=color)
    return padded_image

# Função para extrair características usando VGG16 e preencher imagens
def extract_features(generator, model, target_height, target_width):
    features = []
    labels = []
    for inputs_batch, labels_batch in generator:
        # Aplicar preenchimento a cada imagem no lote
        padded_batch = np.array([pad_image(image, target_height, target_width) for image in inputs_batch])

        # Extrair características usando o modelo VGG16
        features_batch = model.predict(padded_batch)
        features.append(features_batch)
        labels.append(labels_batch)

        if len(features) * generator.batch_size >= generator.samples:
            break
    return np.vstack(features), np.concatenate(labels)

X_train, y_train = extract_features(train_generator, model, img_height, img_width)
X_val, y_val = extract_features(val_generator, model, img_height, img_width)
X_test, y_test = extract_features(test_generator, model, img_height, img_width)

# Padronizar dados
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Reduzir dimensionalidade com PCA para treinamento e avaliação
n_components = 50  # Número de componentes para treinar os classificadores
pca = PCA(n_components=n_components)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

# Reduzir a dimensionalidade para 2 para visualização
pca_2d = PCA(n_components=2)
X_train_pca_2d = pca_2d.fit_transform(X_train_scaled)
X_test_pca_2d = pca_2d.transform(X_test_scaled)

# Matriz de covariância para Mahalanobis
cov_matrix_pca = np.cov(X_train_pca, rowvar=False)
inv_cov_matrix_pca = np.linalg.inv(cov_matrix_pca)

cov_matrix_pca_2d = np.cov(X_train_pca_2d, rowvar=False)
inv_cov_matrix_pca_2d = np.linalg.inv(cov_matrix_pca_2d)

# Classificadores
metrics = ['euclidean', 'manhattan', 'mahalanobis']
k_values = [3, 5]
results = []

for metric in metrics:
    for k in k_values:
        if metric == 'mahalanobis':
            centroid_classifier = NearestCentroidMahalanobis()
            centroid_classifier.fit(X_train_pca, y_train)
        else:
            centroid_classifier = NearestCentroid(metric=metric)
            centroid_classifier.fit(X_train_pca, y_train)

        if metric == 'mahalanobis':
            knn_classifier = KNeighborsClassifier(n_neighbors=k, metric=metric, metric_params={'VI': inv_cov_matrix_pca})
        else:
            knn_classifier = KNeighborsClassifier(n_neighbors=k, metric=metric)
        knn_classifier.fit(X_train_pca, y_train)

        centroid_pred = centroid_classifier.predict(X_test_pca)
        knn_pred = knn_classifier.predict(X_test_pca)

        centroid_cm = confusion_matrix(y_test.astype(np.int_), centroid_pred)
        knn_cm = confusion_matrix(y_test.astype(np.int_), knn_pred)

        centroid_accuracy = np.trace(centroid_cm) / np.sum(centroid_cm)
        knn_accuracy = np.trace(knn_cm) / np.sum(knn_cm)

        results.append((metric, k, centroid_accuracy, knn_accuracy, centroid_cm, knn_cm, centroid_classifier, knn_classifier, X_train_pca, y_train))

        # Plotar regiões de decisão e matrizes de confusão
        plt.figure(figsize=(14, 12))

        # Treinar novos classificadores em PCA 2D para visualização
        if metric == 'mahalanobis':
            centroid_classifier_2d = NearestCentroidMahalanobis()
            knn_classifier_2d = KNeighborsClassifier(n_neighbors=k, metric=metric, metric_params={'VI': inv_cov_matrix_pca_2d})
        else:
            centroid_classifier_2d = NearestCentroid(metric=metric)
            knn_classifier_2d = KNeighborsClassifier(n_neighbors=k, metric=metric)

        centroid_classifier_2d.fit(X_train_pca_2d, y_train)
        knn_classifier_2d.fit(X_train_pca_2d, y_train)

        # Plotar regiões de decisão para NearestCentroid
        plt.subplot(2, 2, 2)
        plot_decision_regions(X_test_pca_2d, y_test.astype(np.int_), clf=centroid_classifier_2d, legend=2)
        plt.title(f'NearestCentroid (Métrica: {metric})')

        # Plotar regiões de decisão para k-NN
        plt.subplot(2, 2, 4)
        plot_decision_regions(X_test_pca_2d, y_test.astype(np.int_), clf=knn_classifier_2d, legend=2)
        plt.title(f'KNeighborsClassifier (k: {k}, Métrica: {metric})')

        # Plotar matriz de confusão para NearestCentroid
        plt.subplot(2, 2, 1)
        sns.heatmap(centroid_cm, annot=True, fmt='d', cmap='Reds', cbar=False)
        plt.title(f'Matriz de Confusão - Centroide (Métrica: {metric}, Acurácia: {centroid_accuracy:.2f})')
        plt.xlabel('Previsto')
        plt.ylabel('Real')

        # Plotar matriz de confusão para k-NN
        plt.subplot(2, 2, 3)
        sns.heatmap(knn_cm, annot=True, fmt='d', cmap='Blues', cbar=False)
        plt.title(f'Matriz de Confusão - k-NN (k: {k}, Métrica: {metric}, Acurácia: {knn_accuracy:.2f})')
        plt.xlabel('Previsto')
        plt.ylabel('Real')

        plt.tight_layout()
        plt.show()

# Estatísticas descritivas
centroid_accuracies = [result[2] for result in results]
knn_accuracies = [result[3] for result in results]

centroid_stats = {
    'min': np.min(centroid_accuracies),
    'max': np.max(centroid_accuracies),
    'mean': np.mean(centroid_accuracies),
    'std': np.std(centroid_accuracies)
}

knn_stats = {
    'min': np.min(knn_accuracies),
    'max': np.max(knn_accuracies),
    'mean': np.mean(knn_accuracies),
    'std': np.std(knn_accuracies)
}

print("Estatísticas de desempenho Nearest Centroid:")
print(centroid_stats)

print("\nEstatísticas de desempenho para k-NN:")
print(knn_stats)
```
